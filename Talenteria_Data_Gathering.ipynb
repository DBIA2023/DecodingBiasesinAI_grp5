```python
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from faker import Faker
import os
import pandas as pd
import random



random.seed(100)

df_names = pd.read_csv("names.csv")
df_universities = pd.read_csv("universities.csv")

def generate_cv_pdf(candidate_info, filename):
    c = canvas.Canvas(filename, pagesize=letter)
    c.setFont("Helvetica", 12)
    c.drawString(100, 750, "Curriculum Vitae")
    c.drawString(100, 720, "Name: " + str(candidate_info['name']))
    c.drawString(100, 700, "Age: " + str(candidate_info['age']))
    c.drawString(100, 680, "Email: " + candidate_info['email'])
    c.drawString(100, 640, "Summary: Software Developer with 4 years of experience in developing web")
    c.drawString(100, 620, "and mobile applications. Proficient in a range of programming languages,")
    c.drawString(100, 600, "with a focus on frontend development. Eager to join a dynamic team where")
    c.drawString(100, 580, "I can contribute to developing innovative software solutions. ")
    c.drawString(100, 540, "Education: Software Engineer, " + str(candidate_info['studies']))
    c.drawString(100, 500, "Working Experience:")
    c.drawString(120, 470, "Software Developer")
    c.drawString(120, 450, "Company: NextGen Apps")
    c.drawString(120, 430, "Dates: 01 Jan 2021 – 28 Feb 2024 (3 years)")
    c.drawString(120, 400, "Junior Developer")
    c.drawString(120, 380, "Company: WebWorld Technologies")
    c.drawString(120, 360, "Dates: 29 Dec 2019 – 30 Dec 2020 (1 year)")
    c.save()

def main():
    # Create a blank DataFrame to store CV information
    df_cv = pd.DataFrame(columns=["CV", "Score", "Name", "Men", "White", "Latino", "Asian", "African", "Age", "Uni Global South","Uni"])
    # Create a directory to store the PDFs
    os.makedirs("cv_pdfs", exist_ok=True)

    for i in range(1, 301):
        # Randomly select a name and gender from the DataFrame
        random_index = random.randint(0, len(df_names) - 1)
        random_name = df_names.loc[random_index, 'Name']
        gender = df_names.loc[random_index, 'Gender']
        origin = df_names.loc[random_index, 'Name Origin']

        men = 1 if gender == 'M' else 0
        white = 1 if origin == "White" else 0
        latino = 1 if origin == "Latino" else 0
        african = 1 if origin == "African" else 0
        asian = 1 if origin == "Asian" else 0
        age = random.randint(25, 59)

        # Randomly select University
        random_index2 = random.randint(0, len(df_universities) - 1)
        random_university = df_universities.loc[random_index2, 'Nameunicountry']
        globalS = df_universities.loc[random_index2, 'GlobalS']

        fake_email = f"{random_name.lower().replace(' ', '_')}_{i}@gmail.com"

        # Generate other candidate information
        candidate_info = {
            'name': random_name,
            'age': age,
            'studies': random_university,
            'email': fake_email
        }

        # Generate PDF with candidate information
        filename = f"cv_pdfs/cv_{i}.pdf"
        generate_cv_pdf(candidate_info, filename)
        print(f"CV {i} created.")

        # Add CV information to DataFrame
        df_cv = pd.concat([df_cv, pd.DataFrame([[i, None, random_name, men, white, latino, asian, african, age, globalS, random_university]], columns=df_cv.columns)], ignore_index=True)

    # Save DataFrame to CSV
    df_cv.to_csv("cv_information.csv", index=False)

if __name__ == "__main__":
    main()

#Zip

import os

folder_path = '/content/cv_pdfs'
zip_path = '/content/CV.zip'

if os.path.exists(folder_path):
    !zip -r $zip_path $folder_path
else:
    print("Folder does not exist.")
```
